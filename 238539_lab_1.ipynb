```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
data=pd.read_csv('/content/sample_data/Churn_Modelling.csv')
data
```

```python
data=data[['RowNumber',	'CustomerId',	'Surname',	'CreditScore',	'Geography',	'Gender',	'Age',	'Tenure',	'Balance',	'NumOfProducts',	'HasCrCard','IsActiveMember',	'EstimatedSalary',	'Exited']]
data
```

```python
for col in data.columns:
  data[col]=data[col].fillna(data[col].mode()[0])
```

```python
data = data.fillna(data.mean())
```

```python
print(data.describe())
```

```python
data = (data - data.mean()) / data.std()
```

```python
data = data.apply(lambda x: (x - x.min()) / (x.max() - x.min()))
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.hist(data['CreditScore'], bins=20)
plt.title("CreditScore")
plt.xlabel("Credit Score")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
for i in data['CreditScore']:
    q1=data['CreditScore'].quantile(0.25)
    q3=data['CreditScore'].quantile(0.75)
    iqr=q3-q1
    lower=q1-1.5*iqr
    upper=q3+1.5*iqr
    if i>upper or i<lower:
          data['CreditScore']= data['CreditScore'].replace(i,np.median(data['CreditScore']))
sns.boxplot(data['CreditScore'])
plt.title("Boxplot of CreditScore")
plt.show()
```

```python
q1 = data['CustomerId'].quantile(0.25)
q3 = data['CustomerId'].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

for i in data['CustomerId']:
    if i > upper or i < lower:
        data.loc[(data['CustomerId'] == i), 'CustomerId'] = np.median(data['CustomerId'])

sns.boxplot(data['CustomerId'])
plt.title("Boxplot of CustomerId")
plt.show()

```

```python
for i in data['Tenure']:
    q1 = data['Tenure'].quantile(0.25)
    q3 = data['Tenure'].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    if i > upper or i < lower:
        data['Tenure'] = data['Tenure'].replace(i, np.median(data['Tenure']))

sns.boxplot(data['Tenure'])
plt.title("Boxplot of Tenure")
plt.show()
```

```python
for i in data['Geography_France']:
    q1 = data['Geography_France'].quantile(0.25)
    q3 = data['Geography_France'].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    if i > upper or i < lower:
        data['Geography_France'] = data['Geography_France'].replace(i, np.median(data['Geography_France']))

sns.boxplot(data['Geography_France'])
plt.title("Boxplot of Geography_France")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = data.iloc[:, :-2].values
y = data.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_test
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
